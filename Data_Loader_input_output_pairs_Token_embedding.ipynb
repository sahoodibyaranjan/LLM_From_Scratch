```python
from torch.utils.data import Dataset, DataLoader

class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):  #max_length is context_size and stride is how much slide to be done
        self.input_ids = []
        self.target_ids = []
        
        #tokenzation of entire text
        token_ids = tokenizer.encode(txt,allowed_special={'<|endoftext|>'})
        
        for i in range(0,len(token_ids)-max_length,stride):
            input_chunck = token_ids[i:i + max_length]
            target_chunck = token_ids[i+1:i + max_length+1]
            self.input_ids.append(torch.tensor(input_chunck))
            self.target_ids.append(torch.tensor(target_chunck))
    
    def __len__(self):
        return len(self.input_ids)
    
    
    def __getitem__(self,idx):
        return self.input_ids[idx], self.target_ids[idx]
            
```

```python
import tiktoken

def create_dataloader_v1(txt,batch_size=8, max_length=256, 
                      stride=128, shuffle=True, 
                      drop_last=True,num_workers=0):  #batch_size is the number of batches the model processes at once before updating its parameter
                                                    # max_length is the context window, gpt can handle 256 context window
                                                    # stride is at a time how much shift you want to do
                                                    # shuffle is randomly shuffle data before batching the data if it is set to true
                                                    # drop_last is used to control whether you want to drop the incomplete batch while batching the dataset
                                                    # num_worker controls the number of parallel worker process used for data loading
    #Initialization of tokenizer
    tokenizer = tiktoken.get_encoding("gpt2")
    
    #Creation of dataset
    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)
    
    #create Dataloader
    dataloader = DataLoader(
        dataset,
        batch_size=batch_size,
        shuffle=shuffle,
        drop_last=drop_last,
        num_workers=num_workers
    )
    return dataloader
```

```python
context_length = max_length

positional_embedding_layer = torch.nn.Embedding(context_length,out_dim)

pos_embedding = positional_embedding_layer(torch.arange(max_length))

print(positional_embedding_layer.weight), print(pos_embedding.shape)
```

```python
final_input_embedding = token_embedding + pos_embedding

print(final_input_embedding), print(final_input_embedding.shape)
```
