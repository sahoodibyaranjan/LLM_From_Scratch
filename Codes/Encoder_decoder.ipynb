# Simple standard function for Encoding and Decoding

```python
class SimpleTokenizerV1:       #please note this implementation will fail if any of the token is out of vocab list
    def __init__(self,vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}
    def encode(self,text):
        preprocessed = re.split(r'([,.:;?_!"()\']|\s)',text) #tokenizing the input text
        preprocessed = [item.strip() for item in preprocessed if item.strip()] #Removing white space to reduce the memory utilization and compute time
        ids = [self.str_to_int[s] for s in preprocessed] #Converting tokens to token ids
        return ids
    def decode(self,ids):
        text = " ".join(self.int_to_str[i] for i in ids) #converting back tokens ids to tokens
        text = re.sub(r'\s+([,.:;?_!"()\'])',r'\1',text) #removing extra space before punctuations
        return text
```

# Handling Unknow tokens and end of text

```python


preprocessed = re.split(r'([,.:;?_!"()\']|\s)',text)
preprocessed = [item.strip() for item in preprocessed if item.strip()]

all_tokens = sorted(set(preprocessed))
all_tokens.extend(["<|endoftext|>","<|unk|>"])

vocab = {token:integer for integer,token in enumerate(all_tokens)}

print(vocab)
```

```python
class SimpleTokenizerV2:       #please note this implementation handles unknown tokens
    def __init__(self,vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}
    def encode(self,text):
        preprocessed = re.split(r'([,.:;?_!"()\']|\s)',text) #tokenizing the input text
        preprocessed = [item.strip() for item in preprocessed if item.strip()] #Removing white space to reduce the memory utilization and compute time
        preprocessed = [item if item in self.str_to_int else "<|unk|>" for item in preprocessed] #handling the unkown tokens
        ids = [self.str_to_int[s] for s in preprocessed] #Converting tokens to token ids
        return ids
    def decode(self,ids):
        text = " ".join(self.int_to_str[i] for i in ids) #converting back tokens ids to tokens
        text = re.sub(r'\s+([,.:;?_!"()\'])',r'\1',text) #removing extra space before punctuations
        return text
```
